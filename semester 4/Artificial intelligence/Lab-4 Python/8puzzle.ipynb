```python
import numpy as np 
from queue import Queue
from matplotlib import pyplot as plt
import heapq
from colorama import Fore, Back, Style
import random
```

```python
canIterate=10000
initialState = [[ [ 1, 4, 5 ],
            [ 7, 3, 2 ],
            [ 8, 0, 6 ] ], 0]
```

```python
goalState = [ [ 1, 2, 3 ],
          [ 4, 5, 6 ],
          [ 7, 8, 0 ] ]
```

```python
def getInd(value):
    indexes=[]
    for row in range(3):
        for col in range(3):
            if(goalState[row][col]==value):
                indexes.append(row)
                indexes.append(col)
                return indexes
                  
```

```python
def moveUp(state,row,col):
    temp=state[row][col]
    state[row][col]=state[row-1][col]
    state[row-1][col]=temp
```

```python
def moveDown(state,row,col):
    temp=state[row][col]
    state[row][col]=state[row+1][col]
    state[row+1][col]=temp 
```

```python
def moveLeft(state,row,col):
    temp=state[row][col]
    state[row][col]=state[row][col-1]
    state[row][col-1]=temp
```

```python
def moveRight(state, row, col):
    temp = state[row][col]
    state[row][col] = state[row][col+1]
    state[row][col+1] = temp
```

```python
def copying(state):
    newState = [[], state[1]]

    for row in state[0]:
        r = []
        for col in row:
            r.append(col)
        newState[0].append(r)

    return newState
```

```python
def expansionSpace(state):
    possibleStates=[]
    for row in range(3):
        for col in range(3):
            getState=state[0]
            if(getState[row][col]==0):
                if(row==0 and col==0):
                    changedState=copying(state)
                    moveDown(changedState[0],row,col)
                    changedState[1] += 1           
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveRight(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    return possibleStates
                elif(row==0 and col==1):
                    changedState=copying(state)
                    moveDown(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveRight(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveLeft(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    return possibleStates
                elif(row==0 and col==2):
                    changedState=copying(state)
                    moveDown(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveLeft(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    return possibleStates
                elif(row==1 and col==0):
                    changedState=copying(state)
                    moveDown(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)                   
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    moveRight(changedState[0],row,col)
                    changedState=copying(state)
                    changedState[1] += 1
                    moveUp(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    return possibleStates 
                elif(row==1 and col==1):
                    changedState=copying(state)
                    moveDown(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveRight(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveLeft(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveUp(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    return possibleStates 
                elif(row==1 and col==2):
                    changedState=copying(state)
                    moveDown(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveLeft(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveUp(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    return possibleStates  
                elif(row==2 and col==0):
                    changedState=copying(state)
                    moveRight(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveUp(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    return possibleStates 
                elif(row==2 and col==1):
                    changedState=copying(state)
                    moveRight(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveLeft(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveUp(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    return possibleStates 
                elif(row==2 and col==2):
                    changedState=copying(state)
                    moveLeft(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    changedState=copying(state)
                    moveUp(changedState[0],row,col)
                    changedState[1] += 1
                    possibleStates.append(changedState)
                    return possibleStates 
```

```python
def heuristic(state):
    manhattenDist=0
    for row in range(3):
        for col in range(3):
            if(state[row][col]==0):
                continue
            else:
                indexes=getInd(state[row][col])
                manhattenDist += abs(row-indexes[0])+abs(col-indexes[1])
    return manhattenDist
```

```python
def Astar(state):
    pque = [[0, state]]
    heapq.heapify(pque)
    iteration=0
    while(len(pque)>0):
        currentPair=heapq.heappop(pque)
        currentState=currentPair[1][0]
        iteration+=1
        if(heuristic(currentState)==0):
            print(Fore.GREEN+"Solved Puzzle!")
            print(Fore.BLUE+"iterations -> "+ str(iteration) + " cost-> "+str(currentPair[1][1]))
            print(Fore.RED+"initialState->" + str(initialState[0]))
            print(Fore.CYAN +"finalState-> "+str(currentState))
            break
        if(iteration==canIterate):
            print(Fore.RED+"Oops! can't solve puzzle")
            print(Fore.BLUE+"iterations -> "+ str(iteration) + " cost-> "+str(currentPair[1][1]))
            print(Fore.RED+"initialState->" + str(initialState[0]))
            print(Fore.CYAN +"finalState-> "+str(currentState))
            break
        allStates=expansionSpace(currentPair[1])
        for newState in allStates:
            State=newState[0]
            Cost=newState[1]
            heapq.heappush(pque,[heuristic(State)+Cost,list(newState)])


        
```
