```python
import math
import random
import copy
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load our dataset
x_train = np.random.uniform(1,3,size= 6)         # features
y_train = 0.5 * x_train + 1.7
```

```python
plt.scatter(x_train, y_train)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python

# function to  compute the cost
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i]) ** 2
    total_cost = (1 / (2*m))*cost
    return total_cost
```

```python
# compute gradient descent
def compute_gradient_descent(x, y, w, b):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = (f_wb - y[i])
        dj_db += dj_db_i
        dj_dw += dj_dw_i
    dj_dw = dj_dw/m
    dj_db = dj_db/m

    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost_function, gradient_fuction):
  # an array to store cost j and w's at each iteration primarily for graph
  j_history = []
  p_history = []
  b = b_in
  w = w_in

  plt.scatter(x_train, y_train, label='Data points',color = 'red')
  for i in range(num_iters):
    # calculate gradient and update the parameters using gradient_descent function
    dj_dw, dj_db = gradient_fuction(x, y, w, b)
    
    print(f"cost fn value {i} : {cost_function(x,y,w,b)}")
    # print(cost_function(x,y,w,b))

    # update parameters
    b = b - alpha * dj_db
    w = w - alpha * dj_dw
    predictions = b + w * x_train
    plt.plot(x_train, predictions, label=f"Best-fit line after {i}th epochs")

  plt.xlabel('X')
  plt.ylabel('Y')
  plt.legend()
  plt.show()

  return w, b, j_history, p_history  # return w and J,w history for graphing

# initialize parameters
w_init = 0
b_init = 0

# some gradient descent settings
iterations = 3
tmp_alpha = 1.0e-1
# run gradient descent
w_final, b_final, J_hist, p_hist = gradient_descent(
    x_train, y_train, w_init, b_init, tmp_alpha, iterations, compute_cost, compute_gradient_descent)
```

```python
print("Predicted line : ")
print(" y = " + str(b_final) + " + " + str(w_final) + " * x")
predictions = b_final + w_final * x_train
```

```python
plt.scatter(x_train, y_train, label='Data points')
plt.plot(x_train, predictions, color='red', label='Best-fit line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
