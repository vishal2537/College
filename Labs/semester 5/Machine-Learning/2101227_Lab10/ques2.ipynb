```python
import numpy as np
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt


X, y = make_classification(n_samples=1000, n_features=5, n_classes=2, n_clusters_per_class=2, n_informative=3, random_state=0)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print("Data Set")
print(X)
print("Target")
print(y)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python

def forward_propagation(X):
    # Calculate the hidden layer activations
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)

    # Calculate the output layer activations
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    return a2 
```

```python
num_samples = 1000
num_features = 5

input_size = num_features
hidden_size = 10
output_size = 2

# Weights and biases for the input to hidden layer
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))

# Weights and biases for the hidden to output layer
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))
```

```python
output = forward_propagation(X)

# Calculate the cross-entropy loss
epsilon = 1e-15  # Small constant to prevent division by zero
loss = -np.mean(y * np.log(output[:, 0] + epsilon) + (1 - y) * np.log(output[:, 1] + epsilon))

threshold = 0
# Convert probabilities to binary class labels
predicted_labels = (output[:, 1] >= threshold).astype(int)  # Assuming class 1 is the positive class

# Calculate accuracy
accuracy = np.mean(predicted_labels == y) * 100

print("Predicted Class Labels (Threshold 0):")
print(predicted_labels)  # Binary class labels (0 or 1)

print("Accuracy:", accuracy, "%")

print("Cross-Entropy Loss:", loss)
print("Accuracy:", accuracy, "%")

print("Output / Predictions")
print(output)
```
