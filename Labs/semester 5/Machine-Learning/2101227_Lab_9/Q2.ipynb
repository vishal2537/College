```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
data = pd.read_csv("credit_card_customer_data.csv")
scale = StandardScaler()
data = scale.fit_transform(data)
```

```python

# Initialize AgglomerativeClustering with linkage='single'
single_model = AgglomerativeClustering(
    n_clusters=None, linkage='single', distance_threshold=0)
single_model.fit(data)

# Initialize AgglomerativeClustering with linkage='complete'
complete_model = AgglomerativeClustering(
    n_clusters=None, linkage='complete', distance_threshold=0)
complete_model.fit(data)

# Initialize AgglomerativeClustering with linkage='average'
average_model = AgglomerativeClustering(
    n_clusters=None, linkage='average', distance_threshold=0)
average_model.fit(data)

# Initialize AgglomerativeClustering with linkage='ward' (ward linkage is similar to centroid linkage)
ward_model = AgglomerativeClustering(
    n_clusters=None, linkage='ward', distance_threshold=0)
ward_model.fit(data)
```

```python

single_linkage_matrix = linkage(data, method='single')
print(single_linkage_matrix)
```

```python
average_linkage_matrix = linkage(data, method='average')
print(average_linkage_matrix)
```

```python
complete_linkage_matrix = linkage(data, method='complete')
print(complete_linkage_matrix)
```

```python
ward_linkage_matrix = linkage(data, method='ward')
print(ward_linkage_matrix)
```

```python

# Plot dendrograms for each linkage function
def plot_dendrogram(linkage_matrix, title):
    dendrogram(linkage_matrix, no_labels=True)
    plt.title(title)
    plt.show()
```

```python


plot_dendrogram(single_linkage_matrix, "Single Linkage Dendrogram")
single_labels = single_model.labels_
```

```python

plot_dendrogram(complete_linkage_matrix, "Complete Linkage Dendrogram")
complete_labels = complete_model.labels_
```

```python

plot_dendrogram(average_linkage_matrix, "Average Linkage Dendrogram")
average_labels = average_model.labels_
```

```python

plot_dendrogram(ward_linkage_matrix,
                "Ward Linkage Dendrogram (similar to Centroid)")
ward_labels = ward_model.labels_
```
