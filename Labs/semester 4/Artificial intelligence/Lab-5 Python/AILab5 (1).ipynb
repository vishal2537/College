# genetic algorithm


Given a set of non-negative integers and a value sum, determine if there is a subset of the
given set with a sum equal to a given sum.

```python
nums = [1,1,8,1,2,3,6]
threshold = 6

def sum(list):
    sumv = 0
    for x in list:
        sumv += x;
    return sumv
    

flag = True
i=0
while(flag and i<3):
#selection
    # randomly select elements from nums
    num_elements = random.randint(1, len(nums))
    random_nums = random.sample(nums, num_elements)

# fitness function
# finding a subset that has difference of subset sum <= threshold/4

    subsets = []

    def find_subsets(random_nums, threshold):
        """ This function takes a list of non-negative integers and a threshold value.
        returns list of all subsets of the input list that have a sum greater than or equal to the threshold value,
        the subset sum and the difference between the subset sum and the threshold value """
        result = []
    
        def create_subset(current_sum, current_index, current_subset):
            if current_sum >= threshold:
                subset_sum = sum(current_subset)
                diff = abs(subset_sum - threshold)
                if diff <= threshold/4:
                    result.append((current_subset, subset_sum, diff))
                
            for i in range(current_index, len(nums)):
                create_subset(current_sum + nums[i], i + 1, current_subset + [nums[i]])
    
        create_subset(0, 0, [])
        return result

    result = find_subsets(random_nums,threshold)


#selection

# sort the result list by the difference between the subset sum and the threshold value
    result = sorted(result, key=lambda x: x[2])

    print("\nSubsets after selection:")
    for subset, subset_sum, diff in result:
        print(subset, "Sum:", subset_sum, "Diff:", diff)
    

# crossover
# select the subset of minimum length and delete the other(s) with maximum length
        deln = len(result)/6
        delcnt = 0
        if len(result) > 1 and delcnt<deln:
            max_length_subsets = max(result, key=lambda x: len(x[0]))
            result.pop(result.index(max_length_subsets))
            delcnt = delcnt+1
        # delete other subset(s) with maximum length
    print("\nSubsets after crossover:")
    for subset, subset_sum, diff in result:
        print(subset, "Sum:", subset_sum, "Diff:", diff)

# mutation
# swap elements of subsets with the least diff and max diff
    if len(result) > 1:
        result[0][0][0], result[-1][0][0] = result[-1][0][0], result[0][0][0]
        result[0] = (result[0][0], sum(result[0][0]), abs(sum(result[0][0]) - threshold))
        result[-1] = (result[-1][0], sum(result[-1][0]), abs(sum(result[-1][0]) - threshold))
        print("\nSubsets after (mutation)swapping:")
        for subset, subset_sum, diff in result:
            print(subset, "Sum:", subset_sum, "Diff:", diff)

# select the subset with a subset sum equal to the threshold, if exists
    for subset, subset_sum, diff in result:
        if subset_sum == threshold:
            result = [(subset, subset_sum, diff)]
            break

    print("\nSubsets after final selection:")
    for subset, subset_sum, diff in result:
        print(subset, "Sum:", subset_sum, "Diff:", diff)
    

#optimal 
    if len(result) >= 1:
        print("Subset(s) found! \n optimal solution")
        flag = False
i = i+1


```

```python
U = {1, 2, 3, 4, 5}
S = [{1}, {2, 3}, {4}, {1, 2}, {1, 3}, {1, 2, 3}, {1, 2, 4}, {1, 3, 5}, {1, 2, 3, 5}, {1, 2, 3, 5},{1 ,2, 5},
     {2, 3, 5}, {2, 3, 4, 5}, {4, 5}, {3, 4, 5}]
n=10
i=0
flag = True
while flag and i<3:
# initial population
    print("initial population generated using subsets :")
    initial_pop = random.sample(S, n)
    print(initial_pop)

# Generate all combinations of 3 sets from S
    combinations = itertools.combinations(S, 3)

# Select 10 random combinations and store them in a list
    combo_list = []
    all_combos = list(combinations)
    for i in range(10):
        random_combo = random.choice(all_combos)
        combo_list.append(random_combo)
        all_combos.remove(random_combo)

# Print the list of combinations
    print("\nInitial population : ")
    for combo in combo_list:
        print(combo)
    
# print(combo_list)

# fitness function
    def fitnessfunc(combo_list):
        for combo in combo_list:
            combo_weight = 0
            unique_nums = set()
            for subset in combo:
                unique_nums.update(subset)
            num_unique = len(unique_nums)
            if num_unique == 1:
                combo_weight = 1
            elif num_unique == 5:
                combo_weight = 5
            else:
                combo_weight = num_unique
            combo_list[combo_list.index(combo)] = (combo, combo_weight)

# Print the list of combinations with weights
    fitnessfunc(combo_list)
    print("\nPopulation with Fitness value :")
    for combo , combo_weight in combo_list:
        print(combo,combo_weight)



# selection
    print("\nPopulation after selection : ")
    max_weight = max(combo_list, key=lambda x: x[1])[1]
    combo_list = [combo for combo in combo_list if combo[1] == max_weight]
    for combo , combo_weight in combo_list:
        print(combo,combo_weight)

# crossover
    max_combo_length = 0
    max_combo = None

    for combo in combo_list:
        max_set_length = max(len(s) for s in combo[0])
        if max_set_length > max_combo_length:
            max_combo_length = max_set_length
            max_combo = combo

    combo_list.remove(max_combo)
    print("\nPopulation after crossover : ")
    for combo , combo_weight in combo_list:
        print(combo,combo_weight)
    
# mutation
    # check if subsets can be removed without changing combo_weight
    for combo, combo_weight in combo_list:
        for subset in combo:
            subset_weight = len(subset)
            if subset_weight == combo_weight:
                temp_combo = list(combo)
                temp_combo.remove(subset)
                temp_combo_weight = 0
                unique_nums = set()
                for s in temp_combo:
                    unique_nums.update(s)
                num_unique = len(unique_nums)
                if num_unique == 1:
                    temp_combo_weight = 1
                elif num_unique == 5:
                    temp_combo_weight = 5
                else:
                    temp_combo_weight = num_unique
                if temp_combo_weight == combo_weight:
                    combo_list[combo_list.index((combo, combo_weight))] = (tuple(temp_combo), temp_combo_weight)
                    break

#if possible that a subset can be removed from the list of combo_list and have weight 5 update changes
    print("\nPopulation after mutation : ")
    for combo , combo_weight in combo_list:
        print(combo,combo_weight)
    
    
#optimal
    # Choose combo with minimum number of sets
    combo_list = sorted(combo_list, key=lambda x: (len(x[0]), -max(len(s) for s in x[0])))

# Select combo with minimum number of sets and maximum subset length
    min_sets = len(combo_list[0][0])
    max_subset_len = max(len(s) for s in combo_list[0][0])
    selected_combo = None

    for combo, combo_weight in combo_list:
        if len(combo) == min_sets and max(len(s) for s in combo) == max_subset_len:
            selected_combo = combo
            break

if selected_combo:
    print(f"\nSelected combo with {min_sets} sets and maximum subset length of {max_subset_len}: {selected_combo}")
    flag = False

else:
    print("\nNo suitable combo found")
    i=i+1




```

Given a knapsack with a weight limit of W, a collection of n items x1, x2, x3, ..., xn with
values v1, v2, v3, ..., vn and weights w1, w2, w3, ..., wn, the knapsack problem is defined as
the optimization problem:

M aximize Xn
i=1
xivi

s.t. Xn
i=1
xiwi ≤ W

xi ∈ {0, 1}, ∀i = 1, 2, .., n

```python
# Knapsack problem parameters
W = 50
n = 10
values = [10, 5, 15, 7, 6, 18, 3, 2, 12, 4]
weights = [7, 4, 10, 5, 4, 12, 2, 1, 6, 3]

# Genetic algorithm parameters
population_size = 50
mutation_rate = 0.01
tournament_size = 3
num_generations = 100

def evaluate(individual):
    total_value = sum(values[i] for i in range(n) if individual[i] == 1)
    total_weight = sum(weights[i] for i in range(n) if individual[i] == 1)
    if total_weight > W:
        return 0
    else:
        return total_value

def select(population):
    tournament = random.sample(population, tournament_size)
    tournament.sort(key=lambda x: evaluate(x), reverse=True)
    return tournament[0]

def crossover(parent1, parent2):
    crossover_point = random.randint(1, n-1)
    offspring1 = parent1[:crossover_point] + parent2[crossover_point:]
    offspring2 = parent2[:crossover_point] + parent1[crossover_point:]
    return offspring1, offspring2

def mutate(individual):
    for i in range(n):
        if random.random() < mutation_rate:
            individual[i] = 1 - individual[i]
    return individual

# Create initial population
population = [ [random.randint(0, 1) for j in range(n)] for i in range(population_size) ]

# Main genetic algorithm loop
for generation in range(num_generations):
    # Evaluate fitness of population
    fitness_values = [evaluate(individual) for individual in population]
    max_fitness = max(fitness_values)
    average_fitness = sum(fitness_values) / population_size
    print(f"Generation {generation+1}: max fitness = {max_fitness}, average fitness = {average_fitness}")
    
    # Select parents
    parents = [select(population) for i in range(population_size)]
    
    # Create offspring through crossover
    offspring = []
    for i in range(0, population_size, 2):
        offspring1, offspring2 = crossover(parents[i], parents[i+1])
        offspring.append(offspring1)
        offspring.append(offspring2)
    
    # Mutate offspring
    mutated_offspring = [mutate(individual) for individual in offspring]
    
    # Replace population with offspring
    population = mutated_offspring

# Find best solution in final population
fitness_values = [evaluate(individual) for individual in population]


```
